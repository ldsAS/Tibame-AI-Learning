## 在對話範本中加入圖片

```python
from google import genai
import urllib.request
from PIL import Image
import requests
from io import BytesIO
from google.genai.types import Content, Part, Blob

# [注意!]跟前面的範例不同 Gemini Chat模式的對話只需要傳入圖片的bytes內容，而非圖片物件本身
# 從URL開啟圖片
response = requests.get("https://img.ltn.com.tw/Upload/health/page/800/2022/01/01/phprhI3u1.jpg")
image_bytes = BytesIO(response.content).getvalue()
# img = Image.open(BytesIO(response.content))

# 從檔案開啟圖片
## img = Image.open("C://DATA/test.jpg")
# with open("C://DATA/test.jpg", "rb") as f:
#     image_bytes = f.read()

contents = [
    Content(role="user", parts=[
        Part(text="這是一張圖片，請使用中文來描述它的內容："),
        Part(
            inline_data=Blob(
                mime_type="image/jpeg",
                data=image_bytes
            )
        )    ])
]

# Configure the client
client = genai.Client(api_key=API_KEY)

# Send request with function declarations
response = client.models.generate_content(
    model=model_name, contents=contents
)

# print(response.candidates[0].content)
print(response.text)
```
