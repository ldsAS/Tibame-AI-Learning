```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("../content/Test/vgsales.csv")
data.head()
```

```python
data.shape
```

```python
# 找出有缺失值的欄位
data.isnull().sum()
# 或是用info(需自行比對總數)
# data.info()
```

```python
# 計算欄位的基本統計數據
data.describe()
# data.describe(include="all") # 可看所有欄位(含非值)
```

```python
# 取出 "Publisher" 欄位的資料
data["Publisher"]
```

```python
# 取出 Publisher, Name, Genre 欄位的資料
data[["Publisher", "Name", "Genre"]]
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和
data[['NA_Sales', "EU_Sales", "JP_Sales", "Other_Sales"]].sum(axis=1)
```

```python
# 另一個寫法(使用sum的方式較佳)
data["NA_Sales"]+data["EU_Sales"]+data["JP_Sales"]+data["Other_Sales"]
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和，並存入新的欄位"總和"
data["總和"] = data[['NA_Sales', "EU_Sales", "JP_Sales", "Other_Sales"]].sum(axis=1)
data.head()
```

```python
#找出發行商為Nintendo的遊戲名稱
data.loc[data["Publisher"] == "Nintendo"]
```

```python
#列出所有Wii的遊戲名稱
data.loc[data["Platform"] == "Wii"]
```

```python
#計算所有GBA遊戲的銷售總合
data.loc[data["Platform"] == "GBA"]["Global_Sales"].sum()
```

```python
# 另一個解法(因利用先加總後的結果，可能會導致四捨五入後的總和有差異)
data.loc[data["Platform"] == "GBA"]["總和"].sum()

# 此輸出為Numpy語法，若有其他使用需求可能就轉換型態，此時可用 float()
```

```python
# 再一另解(邏輯與利用總和去加總的結果相同)
data.loc[data["Platform"] == "GBA"][["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum().sum()
```

```python
#計算任天堂運動類遊戲的銷售總量
data.loc[(data.Publisher == "Nintendo") & (data.Genre == "Sports")]["Global_Sales"].sum()

# And : &
# Or : |
```

```python
#列出北美地區銷售量最高的前五名遊戲
data.sort_values(by="NA_Sales", ascending=False).iloc[:5]

# 另一解法
# data.sort_values(by="NA_Sales", ascending=False).head(5)
```

```python
#列出北美與歐洲地區銷售量總和最高的前五名遊戲

# argsort()傳出來的是排序後的索引(由小排到大)
x = data[["NA_Sales", "EU_Sales"]].sum(axis=1).argsort()
data.iloc[x[-5:]]
```

```python
# 測試另外的方式
data["NA&EU銷售額"] = data[['NA_Sales', "EU_Sales"]].sum(axis=1)
data.sort_values(by="NA&EU銷售額", ascending=False).iloc[:5]
```

```python
# 可使用drop刪除多餘欄位
# data.drop('欄位名稱', axis=1, inplace=True)
```

```python
#找出總共有多少種不同的平台
PV = data["Platform"].unique()
print(PV)
print(len(PV))
```

```python
#寫程式計算不同平台的總銷售量
PV = data["Platform"].unique()
for Name in PV:
    x = (data.Platform == Name)
    s = data[x]["Global_Sales"].sum()
    print(f"{Name} : {s}")
```

```python
# 使用GroupBy來完成上述的工作
data.groupby("Platform")["Global_Sales"].sum()
```
