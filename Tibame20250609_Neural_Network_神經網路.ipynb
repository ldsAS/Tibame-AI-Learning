# 類神經網路 (Neural Network)

類神經網路是一種受人類大腦神經元結構啟發的數學模型，用於進行模式識別、預測、分類等任務。它由一層層的「神經元」組成，這些神經元之間通過權重相連，模擬人腦神經元之間的信息傳遞與處理過程。

## 基本結構

1. **輸入層 (Input Layer)**：接收外部數據，這些數據可以是圖像、文字、音頻等。
2. **隱藏層 (Hidden Layers)**：包含多個神經元，這些神經元進行非線性處理，學習數據的特徵。深度神經網路通常會有多層隱藏層。
3. **輸出層 (Output Layer)**：將隱藏層的處理結果轉化為預測或分類的結果。

每個神經元通過 **激勵函數 (Activation Function)** 對接收到的信號進行轉換，以便引入非線性因素，使網路能夠處理更複雜的問題。

## 訓練過程

類神經網路的訓練過程主要是通過 **反向傳播 (Backpropagation)** 來調整神經元間的權重，使網路能夠減少預測錯誤。訓練過程中，使用 **梯度下降** 或其變種來最小化損失函數，從而使模型不斷提高準確度。

## 應用領域

* 圖像處理
    - **物體識別**：自動識別圖像中的物體，常見於人臉識別、交通標誌識別等。
    - **影像分割**：將圖像分割成多個區域，常見於醫學影像分析。
* 語音處理
    - **語音識別**：將語音轉換為文字，應用於語音助手、語音指令等。
    - **語音合成**：將文字轉換為語音，常見於導航系統和語音助手中。
* 自然語言處理 (NLP)
    - **機器翻譯**：將一種語言翻譯成另一種語言。
    - **情感分析**：分析文本中的情感趨勢，用於社交媒體監控、客服等領域。
    - **問答系統**：根據用戶提問，自動生成回答。
* 金融
    - **股票市場預測**：預測股市走勢，進行智能投資決策。
    - **信用評分**：根據用戶的信用歷史來判斷是否能夠授予貸款。
* 自駕車
    類神經網路可用於車輛的自動駕駛，協助進行路況分析、障礙物識別、路徑規劃等。
* 醫學
    - **疾病診斷**：分析病人的檢查數據（如影像、基因數據等）來輔助診斷疾病。
    - **藥物研發**：使用神經網絡進行藥物的結構預測或藥效評估。

## 優點

- **強大的學習能力**  
   類神經網路能夠從大量數據中自動學習特徵，並且能處理複雜的非線性關係，這使得它在許多領域（如語音識別、圖像處理、自然語言處理等）表現出色。
- **適應性強**  
   類神經網路可以隨著新數據的到來進行更新和調整，適應環境的變化。
- **並行處理能力**  
   類神經網路可以同時處理大量數據，適合用於大規模計算和處理。
- **強大的泛化能力**  
   當訓練得當時，類神經網路能夠對未見過的數據做出合理的預測或分類。
- **應用範圍廣泛**  
   類神經網路已經在許多領域取得成功，如自駕車、醫學診斷、金融風險預測等，並且隨著技術進步，它的應用範圍還在不斷擴大。

## 缺點

- **需要大量的數據**  
   類神經網路通常需要大量的標註數據來進行有效的訓練，這在某些情況下可能是個挑戰，特別是在數據稀缺的領域。
- **計算資源要求高**  
   訓練深度神經網路需要大量的計算資源，特別是在處理大規模數據集時，往往需要強大的硬體支持，如GPU加速。
- **黑箱問題**  
   類神經網路的運作過程較為複雜，難以解釋其內部的決策過程。這使得它在某些需要可解釋性的應用中（如醫學診斷、法律等）存在挑戰。
- **容易過擬合**  
   如果神經網路訓練不當，可能會導致過擬合，即模型在訓練數據上表現很好，但對新數據的預測能力差。
- **需要精細的超參數調整**  
   神經網路的性能對超參數（如學習率、層數、神經元數量等）非常敏感，選擇合適的超參數往往需要經驗或大量的試驗。
- **訓練時間長**  
   由於需要處理大量數據和調整網絡權重，神經網路的訓練過程通常較為漫長，這可能會影響開發和部署的效率。

# 神經元 Neuron

```python
# 這裡先產生一份測試用的資料
import numpy as np
import matplotlib.pyplot as plt
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100, 1)
# 將測試資料畫出來
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# 使用線性回歸來預測
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

## 類神經網路的梯度下降與損失函式

類神經網路是一種模仿人腦神經結構的數學模型，由多層神經元組成。這些神經元接收輸入訊號並經過激活函式處理後，將結果傳遞給下一層神經元。類神經網路的目的是通過學習一組最佳的參數（權重和偏差）來進行預測或分類。

### 損失函式 (Loss Function)
損失函式是衡量模型預測結果與實際結果之間誤差的指標。在神經網路中，模型的輸出是基於權重和偏差計算出來的，但這些預測可能與真實標籤不完全匹配。損失函式的目的是量化這個誤差。常見的損失函式有：
- **均方誤差 (Mean Squared Error, MSE)**：用於回歸問題，計算預測值與真實值之間差異的平方平均值。
- **交叉熵損失 (Cross-Entropy Loss)**：用於分類問題，衡量預測類別與真實標籤之間的差距。
損失函式的數值越小，表示模型的預測越接近真實值，反之則表示預測誤差越大。


## 損失(loss)函示值分布圖
假設要訓練只有一個變數$w_1$的神經元，我們可以用窮舉法把所有$w_1$的值都測試過一遍，最後選擇一個loss計算出來最小的$w_1$值。

```python
w1 = np.arange(-10,10,0.1) #把w1從-10到10所有可能的值都列出來(間格為0.1)
o = w1 * X + 4 #計算不同w1值時的預測
err = o - y #預測值跟實際值的誤差
loss = np.mean(err * err, axis=0) #計算神經元誤差的loss

print('w1 = ', w1[loss.argmin()], ' 的時候loss最小')

#把所有w1值造成誤差的loss值畫出來
plt.plot(w1, loss, "-")
plt.xlabel("$w_1$", fontsize=18)
plt.ylabel("$loss$", fontsize=18)
# plt.axis([0, 2, 0, 15])
plt.show()
```

### 梯度下降 (Gradient Descent)
梯度下降是一種優化算法，用來尋找使損失函式最小化的最佳參數（即神經網路中的權重和偏差）。簡單來說，梯度下降的目標是通過不斷調整參數來減少損失函式的數值。

#### 梯度下降的基本流程：
1. **計算損失函式對每個參數的梯度**，這表示損失函式相對於每個參數變化的方向和速率。
2. **更新參數**，根據梯度的方向來調整參數。這個更新是通過減去梯度乘上一個學習率（learning rate）來實現的。
3. **重複這個過程**，直到損失函式的值收斂到最小值，或達到預定的迭代次數。

具體步驟：
1. **初始化參數**：隨機設定權重和偏差。
2. **計算預測結果**：基於當前的參數進行前向傳播，得到預測結果。
3. **計算損失**：根據損失函式計算預測結果與真實值的差異。
4. **計算梯度**：對損失函式關於權重和偏差求偏導數（即梯度）。
5. **更新參數**：根據梯度和學習率更新權重和偏差。
6. **重複**：重複步驟 2 至 5，直到損失函式收斂。

### 梯度下降的類型：
- **批量梯度下降 (Batch Gradient Descent)**：每次使用整個訓練集來計算梯度並更新參數。這種方法計算量大，但保證收斂。
- **隨機梯度下降 (Stochastic Gradient Descent, SGD)**：每次只使用一個訓練樣本來計算梯度並更新參數。這樣可以更快，但可能會有較大的波動。
- **小批量梯度下降 (Mini-batch Gradient Descent)**：每次使用訓練集的子集（小批量）來計算梯度並更新參數，是一種折衷方法。


## 梯度下降的方程式

在梯度下降中，更新權重的方程式是基於損失函式的梯度進行的。以下是梯度下降的基本方程式：

### 梯度下降更新公式：
$
\theta = \theta - \eta \cdot \nabla_\theta loss(\theta)
$

其中：
- $\theta$ 表示模型的參數（例如權重$w_n$和偏差$b$）。
- $\eta$ 是學習率（learning rate），控制每次更新的步長。
- $\nabla_\theta loss(\theta)$ 是損失函式 $loss(\theta)$ 對參數 $\theta$ 的梯度，表示損失函式隨參數變化的速率和方向。

## 詳細解釋：
- **$\theta$**：這是我們希望優化的參數，通常是神經網路中的權重和偏差。
- **學習率 $\eta$**：它決定了每次參數更新的幅度。如果學習率過大，可能會導致錯過最佳解；如果學習率過小，則可能收斂過慢。
- **$\nabla_\theta loss(\theta)$**：這是梯度，表示損失函式對參數的導數。它告訴我們如何調整參數來減少損失。具體來說，梯度的方向是損失函式增長最快的方向，而梯度下降就是沿著這個方向反向更新參數。

## 梯度下降的步驟：
1. 計算當前參數下的損失函式 $loss(\theta)$。
2. 計算損失函式對參數的梯度 $\nabla_\theta loss(\theta)$。
3. 根據梯度和學習率 $\eta$ 更新參數：$\theta := \theta - \eta \cdot \nabla_\theta loss(\theta)$。
4. 重複步驟 1 至 3，直到損失函式達到最小值或收斂。

這個過程不斷迭代，直到模型找到最佳的參數，從而最小化損失函式。


## 使用 MSE 當 Loss 時的梯度方程式

### MSE 損失函式
給定一組預測值 $ \hat{y} $ 和實際值 $ y $，對於一組訓練樣本 $ (x_i, y_i) $，MSE 損失函式可以表示為：

$
loss(\theta) = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i)^2
$

其中：
- $ m $ 是訓練樣本的數量。
- $ \hat{y}_i $ 是第 $i$ 個樣本的預測值。
- $ y_i $ 是第 $i$ 個樣本的真實值。

### 梯度計算
假設我們的模型是線性回歸模型，預測值 $ \hat{y} $ 可以表示為：
$
\hat{y} = \theta^T x
$
其中 $ x $ 是輸入特徵向量，$ \theta $ 是權重參數向量。

為了最小化 MSE 損失，我們需要對損失函式 $ loss(\theta) $ 關於每個參數 $ \theta_j $ 求偏導數（梯度）。

### 梯度公式
MSE 損失函式對權重 $ \theta_j $ 的偏導數（即梯度）為：

$
\frac{\partial loss(\theta)}{\partial \theta_j} = \frac{2}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i) \cdot x_{ij}
$

其中：
- $ \hat{y}_i = \theta^T x_i $，是第 $ i $ 個樣本的預測值。
- $ x_{ij} $ 是第 $ i $ 個樣本中第 $ j $ 個特徵的值。

### 梯度下降更新公式
根據梯度下降的公式，更新權重 $ \theta_j $ 的方程式為：

$
\theta_j = \theta_j - \eta \cdot \frac{\partial loss(\theta)}{\partial \theta_j}
$

代入梯度公式，得到更新規則：

$
\theta_j = \theta_j - \eta \cdot \frac{2}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i) \cdot x_{ij}
$

## 總結
當使用 MSE 作為損失函式時，梯度下降的更新公式是：
$
\theta_j = \theta_j - \eta \cdot \frac{2}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i) \cdot x_{ij}
$
其中：
- $ \hat{y}_i = \theta^T x_i $
- $ \eta $ 是學習率
- $ m $ 是樣本數量

根據預測值與實際值之間的差異，更新模型參數 $ \theta $，以便使損失函式最小化。


```python
eta = 0.1  # learning rate
n_iterations = 1000
m = 100

theta = np.random.randn(2,1)  # random initialization

for iteration in range(n_iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y) #計算梯度
    theta = theta - eta * gradients #經由梯度來修正w值
```

# 隨機梯度下降 (Stochastic Gradient Descent, SGD)

## SGD 的基本原理
在傳統的 **批量梯度下降** 中，每次迭代都會使用整個訓練集來計算損失函式的梯度，然後更新參數。而在 **隨機梯度下降** 中，每次迭代只會使用一個訓練樣本來計算梯度並更新參數。

## SGD 的優點和缺點

### 優點：
- **高效能**：每次只使用一個樣本來計算梯度，這樣可以大大減少每次迭代的計算量，適用於大規模數據集。
- **更快的收斂**：由於每次更新都基於一個樣本，模型的參數更新速度比批量梯度下降快，並且可以更快地到達一個較好的解。
- **逃脫局部最小值**：SGD 的隨機性可以幫助模型逃脫局部最小值，進而找到更好的全局最小值。

### 缺點：
- **波動性大**：由於每次更新只基於單個樣本，SGD 的更新過程可能會有很大的波動，這會使得損失函式的曲線不夠平滑。
- **收斂速度慢**：儘管每次更新比較快，但由於更新的不穩定，整體收斂過程可能比較慢，需要較多的迭代次數。
- **需要設置學習率**：選擇適當的學習率對於 SGD 的效果非常重要。如果學習率過大，可能會導致不收斂；如果學習率過小，則可能會收斂過慢。

## 4. 優化的 SGD 變體
為了克服 SGD 的缺點，很多優化算法都對 SGD 進行了改進。常見的變體包括：
- **Mini-batch Gradient Descent**：每次迭代使用一小批樣本進行更新，這樣可以在計算效率和穩定性之間找到平衡。
- **Momentum**：通過加入上一輪更新的累積信息來加速收斂，減少波動。
- **Adam**：結合了動量和自適應學習率的方法，是目前深度學習中最常用的優化算法之一。

```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)

n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # random initialization

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    # not shown in the book
            y_predict = X_new_b.dot(theta)           # not shown
            style = "b-" if i > 0 else "r--"         # not shown
            plt.plot(X_new, y_predict, style)        # not shown
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 # not shown

plt.plot(X, y, "b.")                                 # not shown
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, 0, 15])                              # not shown
plt.show()                                           # not shown
```

## 小批量梯度下降 Mini-batch gradient descent

```python
theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
```

```python
theta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)
```

## Perception (神經元 Neuron)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn import metrics

data = pd.read_csv("../content/Iris.csv")
# data.drop('Id',axis=1,inplace=True)

train, test = train_test_split(data, test_size = 0.3, random_state=0)

train_X = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking the training data features
train_y = train['Species'] # output of our training data
test_X = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] # taking test data features
test_y = test['Species']   #output value of test data

model = Perceptron()

model.fit(train_X,train_y) #訓練模組

prediction = model.predict(test_X) #使用訓練好的模組來測試
print('The accuracy is:', metrics.accuracy_score(prediction,test_y))
```

Perceptron is a classification algorithm which shares the same underlying implementation with SGDClassifier. In fact, Perceptron() is equivalent to SGDClassifier(loss="perceptron", eta0=1, learning_rate="constant", penalty=None).

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier()

model.fit(train_X,train_y) #訓練模組

prediction = model.predict(test_X) #使用訓練好的模組來測試
print('The accuracy is:', metrics.accuracy_score(prediction,test_y))
```
