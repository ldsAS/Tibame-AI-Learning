````python
import urllib.request as req
import bs4 as bs
# 0
import pandas as pd

#1 空的list
table = []
for i in range(10):
    page = i + 1
    print("[ 頁數 ", page, " ]")
    url ="https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt".format(page)
    resp = req.urlopen(url)
    content = resp.read()
    html = bs.BeautifulSoup(content)

    rs = html.find_all("div", {"class":"list-rst__body"})
    for r in rs:
        en = r.find("a", {"class":"list-rst__rst-name-target"})
        area_genre = r.find("div", {"class":"list-rst__area-genre"})
        rating = r.find("span", {"class":"list-rst__rating-val"})
        dinner, lunch = r.find_all("span", {"class":"c-rating-v3__val"})
        # process
        en_href = en["href"]
        en_name = en.text
        area_genre_text = area_genre.text.strip()
        area_genre_text_split = area_genre_text.split(" / ")
        # 如果只有一個，地點從缺
        if len(area_genre_text_split) == 1:
            area_text, genre_text = "", area_genre_text_split[0]
        else:
            area_text, genre_Text = area_genre_text_split[0], area_genre_text_split[1]

        rating_text = rating.text

        # print
        # print(en_name)
        # print(rating_text)
        # print(area_text)
        # print(genre_text)
        # print(dinner.text)
        # print(lunch.text)
        # print(en_href)
        # print("-" * 30)

        # 準備好一筆資料(一列)
        data = {
            "評價":rating_text,
            "英文店名":en_name,
            "超連結":en_href,
            "地點":area_text,
            "類別":genre_text,
            "晚餐價格":dinner.text,
            "午餐價格":lunch.text,
        }
        table.append(data)

# 3. 把list/dict轉成pd.DataFrame
df = pd.json_normalize(table)
df.to_csv("tabelog.csv", encoding="utf-8")
````
