## åœ–ç‰‡
gr.Image()åŠŸèƒ½åœ¨æŸäº›ç‰ˆæœ¬çš„Gradioæœƒå¤±æ•ˆç”¢ç”ŸéŒ¯èª¤ï¼Œç›®å‰çš„æ¸¬è©¦æ˜¯gradio==3.48.0å¯ä»¥ç”¨

```python
import numpy as np
import gradio as gr

def sepia(input_img):
    sepia_filter = np.array([
        [0.393, 0.769, 0.189],
        [0.349, 0.686, 0.168],
        [0.272, 0.534, 0.131]
    ])
    sepia_img = input_img.dot(sepia_filter.T)
    sepia_img /= sepia_img.max()
    return sepia_img

demo = gr.Interface(sepia, gr.Image(), "image")
demo.launch()
```

# Blocks & Layout

### gr.Blocks()

```python
import gradio as gr
def update(name):
    return f"Welcome to Gradio, {name}!"

with gr.Blocks() as demo: #å®šç¾©å€å¡Š
    # ä¸‹é¢çš„ç‰©ä»¶æœƒç”±ä¸Šè€Œä¸‹åœ¨ç¶²é ä¸Šå‚ç›´æ–¹å‘çš„å †ç–Š
    gr.Markdown("Start typing below and then click **Run** to see the output.") #é¡¯ç¤ºMarkdownæ ¼å¼çš„ç‰©ä»¶
    with gr.Row(): #ç”¢ç”Ÿä¸€å€‹å·¦å³æ©«å‘å †ç–Šçš„å€å¡Š
        inp = gr.Textbox(placeholder="What is your name?")
        out = gr.Textbox()
    btn = gr.Button("Run") #æŒ‰éµ
    btn.click(
        fn=update, #ç•¶æŒ‰éµæŒ‰ä¸‹æ™‚ï¼Œæ¥æ”¶äº‹ä»¶çš„function
        inputs=inp, #è¦å‚³éçµ¦äº‹ä»¶è™•ç†functionçš„grç‰©ä»¶
        outputs=out) #æŒ‡å®šå°‡äº‹ä»¶è™•ç†functionçš„è¼¸å‡ºå€¼æ”¾å…¥é‚£å€‹grç‰©ä»¶ä¸­

demo.launch() #åŸ·è¡Œå€å¡Š
```

### gr.Accordion()

```python
import gradio as gr

def greet(name, formal):
    if formal:
        return f"Good day to you, {name}."
    else:
        return f"Hey {name}!"

with gr.Blocks() as demo:
    gr.Markdown("## ğŸ‘‹ æ‰“æ‹›å‘¼ Appï¼ˆæœ‰é€²éšé¸é …ï¼‰")

    name_input = gr.Textbox(label="ä½ çš„åå­—")

    with gr.Accordion("é€²éšé¸é …", open=False):
        formal_checkbox = gr.Checkbox(label="ä½¿ç”¨æ­£å¼èªæ°£")

    greet_button = gr.Button("æ‰“æ‹›å‘¼")
    output = gr.Textbox(label="è¼¸å‡º")

    greet_button.click(fn=greet, inputs=[name_input, formal_checkbox], outputs=output)

demo.launch()
```

### gr.Tabs()

```python
import gradio as gr

def greet(name):
    return f"Hello, {name}!"

def farewell(name):
    return f"Goodbye, {name}!"

with gr.Blocks() as demo:
    with gr.Tabs():
        with gr.Tab("Greeting"):
            name_input = gr.Textbox(label="Enter your name")
            greet_button = gr.Button("Greet")
            greet_output = gr.Textbox(label="Output")
            greet_button.click(fn=greet, inputs=name_input, outputs=greet_output)

        with gr.Tab("Farewell"):
            name_input2 = gr.Textbox(label="Enter your name")
            bye_button = gr.Button("Say Goodbye")
            bye_output = gr.Textbox(label="Output")
            bye_button.click(fn=farewell, inputs=name_input2, outputs=bye_output)

demo.launch()
```
